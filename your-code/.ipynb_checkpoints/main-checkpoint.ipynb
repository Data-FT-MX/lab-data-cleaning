#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users_selec_columns = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts_selec_columns = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merge_df = pd.merge(left=users_selec_columns , right=posts_selec_columns, left_on='userId', right_on='userId')
merge_df
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
# Asi podemos ver las graficas y ver si tenemos outliers
for col in merge_df_clean._get_numeric_data():
    print(col)
    print(merge_df_clean[col].plot.box())
    plt.show()
    print('------' * 10)
```

```python
stats = merge_df_clean.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outlier = pd.DataFrame(columns=merge_df_clean.columns)

for col in stats.index:
    iqr = stats.loc[col, 'IQR']
    cutoff = iqr * 1.5
    lower = stats.loc[col, '25%'] - cutoff
    upper = stats.loc[col, '75%'] + cutoff
    result = merge_df_clean[(merge_df_clean[col] < lower) | (merge_df_clean[col] > upper)].copy()
    result['Outlier'] = col
    outlier = outlier.append(result)
    
outlier # Aqui tenemos nuestros valores outliers
```

```python
# Pasamos nuestra data a un CSV

outlier.to_csv('outlier.csv', index=False)
```
